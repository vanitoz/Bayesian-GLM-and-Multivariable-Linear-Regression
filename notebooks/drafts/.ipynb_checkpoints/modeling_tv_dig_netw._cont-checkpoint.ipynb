```python
x = data[[col for col in data.columns if col != 'TOTAL_CONV']].values
y = data.TOTAL_CONV.values
```

```python
# Scaling all data with MinMaxScaler
# scaller = MinMaxScaler()
# data= pd.DataFrame(scaller.fit_transform(data), columns= data.columns)
```

### NegativeBinomial for Likelihood

```python
with pm.Model() as model_neg:

        # Intercept
        alpha = pm.Normal('alpha', mu=y.mean(), sd=10)

        # Slope
        beta = pm.HalfFlat('beta', shape = len(data.columns[:-1]))

        # Error term
        eps = pm.HalfCauchy('eps', 5)

        # Expected value of outcome (ML Regression with vectors)
        mu = alpha + pm.math.dot(x, beta)

        # Likelihood
        conv = pm.NegativeBinomial('conv', 
                                   mu = mu, 
                                   alpha= eps,  
                                   observed=y)

        trace_neg = pm.sample(chains = 4, target_accept = 0.9)
```

```python
az.plot_trace(trace_neg)
plt.show()
```

```python
az.summary(trace_neg, var_names = 'beta').set_index(data.columns[:-1])
```

```python
predictions_neg = ppc_check(model=model_neg, 
                            data = data, 
                            trace= trace_neg, 
                            predictions= True)
```

### ZeroInflated Poisson as Likelihood

```python
with pm.Model() as model_zero_poisson:

        # Intercept
        alpha = pm.Normal('alpha', mu=y.mean(), sd=10)

        # Slope
        beta = pm.HalfFlat('beta', shape = len(data.columns[:-1]))

        # Error term
        eps = pm.HalfCauchy('eps', 5)

        # Expected value of outcome (ML Regression with vectors)
        mu = alpha + pm.math.dot(x, beta)

        # Likelihood
        conv = pm.ZeroInflatedPoisson('conv', 
                                      theta= mu, 
                                      psi= 0.7,  
                                      observed=y)

        trace_zero_poisson = pm.sample(chains = 4, target_accept = 0.9)
```

```python
az.plot_trace(trace_zero_poisson)
plt.show()
```

```python
predictions_zero_poisson = ppc_check(model=model_zero_poisson, 
                            data = data, 
                            trace= trace_zero_poisson, 
                            predictions= True)
```

### Model Comparison

```python
plt.figure(figsize=(10, 5))
ppc_data = [y, 
            predictions_neg['conv'], 
            predictions_zero_poisson['conv'], 
           ]

labels = ['data', 'NegBinominal', 'ZeroPoisson',]
for i, d in enumerate(ppc_data):
    mean = d.mean()
    err = np.percentile(d, [0, 100])
    plt.errorbar(mean, -i, xerr=[[-err[0]], [err[1]]], fmt='o')
    plt.text(mean, -i+0.2, labels[i], ha='center', fontsize=14)
    plt.ylim([-i-0.5, 0.5])
    plt.xlim([-10, 70])
    plt.yticks([])
```
