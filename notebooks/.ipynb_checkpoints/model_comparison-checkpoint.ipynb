```python
x = data[[col for col in data.columns if col != 'TOTAL_CONV']].values
y = data.TOTAL_CONV.values
```

### NegativeBinomial as Likelihood

```python
with pm.Model() as model_neg:

        # Intercept
        alpha = pm.Normal('alpha', mu=y.mean(), sd=10)

        # Slope
        beta = pm.HalfFlat('beta', shape = len(data.columns[:-1]))

        # Error term
        eps = pm.HalfCauchy('eps', 5)

        # Expected value of outcome (ML Regression with vectors)
        mu = alpha + pm.math.dot(x, beta)

        # Likelihood
        conv = pm.NegativeBinomial('conv', 
                                   mu = mu, 
                                   alpha= eps,  
                                   observed=y)

        trace_neg = pm.sample(chains = 4, target_accept = 0.9)
```

```python
az.plot_trace(trace_neg)
plt.show()
```

```python
predictions_neg = ppc_check(model=model_neg, 
                            data = data, 
                            trace= trace_neg, 
                            predictions= True)
```

### ZeroInflated Poisson as Likelihood

```python
with pm.Model() as model_zero_poisson:

        # Intercept
        alpha = pm.Normal('alpha', mu=y.mean(), sd=10)

        # Slope
        beta = pm.HalfFlat('beta', shape = len(data.columns[:-1]))

#         # Error term
#         eps = pm.HalfCauchy('eps', 5)
        psi = pm.Beta('psi', 1, 1)
        # BoundedFlat = pm.Bound(pm.Flat, lower=0.0, upper = 1.0)
        # psi = BoundedFlat('psi')
        
        # Expected value of outcome (ML Regression with vectors)
        mu = alpha + pm.math.dot(x, beta)

        # Likelihood
        conv = pm.ZeroInflatedPoisson('conv', 
                                      theta= mu, 
                                      psi= psi,  
                                      observed=y)

        trace_zero_poisson = pm.sample(chains = 4, target_accept = 0.9)
```

```python
az.plot_trace(trace_zero_poisson)
plt.show()
```

```python
coeff_zip = az.summary(trace_zero_poisson, var_names = 'beta').set_index(data.columns[:-1])['mean']
```

```python
coeff_zip = pd.DataFrame(coeff_zip).rename(columns={'mean':'ZeroPoisson'})
```

```python
predictions_zero_poisson = ppc_check(model=model_zero_poisson, 
                            data = data, 
                            trace= trace_zero_poisson, 
                            predictions= True)
```

```python
with pm.Model() as model_zero_poisson_07:

        # Intercept
        alpha = pm.Normal('alpha', mu=y.mean(), sd=10)

        # Slope
        beta = pm.HalfFlat('beta', shape = len(data.columns[:-1]))
        
        # Expected value of outcome (ML Regression with vectors)
        mu = alpha + pm.math.dot(x, beta)

        # Likelihood
        conv = pm.ZeroInflatedPoisson('conv', 
                                      theta= mu, 
                                      psi= 0.7,  
                                      observed=y)

        trace_zero_poisson_07 = pm.sample(chains = 4, target_accept = 0.9)
```

### ZeroInflatedNegativeBinomial for Likelihood Function

```python
with pm.Model() as model_zero_nb:

        # Intercept
        alpha = pm.Normal('alpha', mu=y.mean(), sd=10)

        # Slope
        beta = pm.HalfFlat('beta', shape = len(data.columns[:-1]))

        # Error term
        eps = pm.Gamma('eps', 1, 0.5)
        
        psi = pm.Beta('psi', 1, 1)

        # Expected value of outcome (ML Regression with vectors)
        mu = alpha + pm.math.dot(x, beta)
        
        conv = pm.ZeroInflatedNegativeBinomial('conv', 
                                              mu = mu,
                                              psi = psi,
                                              alpha = eps,
                                              observed = y)
        
        trace_zero_nb = pm.sample(chains = 4, target_accept = 0.9)
```

```python
predictions_zero_nb = ppc_check(model=model_zero_nb, 
                            data = data, 
                            trace= trace_zero_nb, 
                            predictions= True)
```

### Using Constrain on Zero Negative Binominal Likelihood

```python
with pm.Model() as model_zero_nb_constr:

        # Intercept
        alpha = pm.Normal('alpha', mu=y.mean(), sd=10)

        # Slope
        beta = pm.HalfFlat('beta', shape = len(data.columns[:-1]))

        # Error term
        eps = pm.Gamma('eps', 1, 0.5)
        
        psi = pm.Beta('psi', 1, 1)

        # Expected value of outcome (ML Regression with vectors)
        mu = alpha + pm.math.dot(x, beta)
        
        conv = pm.ZeroInflatedNegativeBinomial('conv', 
                                              mu = mu,
                                              psi = psi,
                                              alpha = eps,
                                              observed = y)
        
        pm.Potential('constraint', tt.switch(conv > 20, -np.inf, 0.))
        
        trace_zero_nb_constr = pm.sample(chains = 4, target_accept = 0.9)
```

```python
az.plot_trace(trace_zero_nb_constr)
plt.show()
```

```python
predictions_zero_nb_constr = ppc_check(model=model_zero_nb_constr, 
                            data = data, 
                            trace= trace_zero_nb_constr, 
                            predictions= True)
```

```python
fast_spp = pm.fast_sample_posterior_predictive(trace=trace_zero_nb_constr, 
                                    model=model_zero_nb_constr, 
                                    samples = 200)
```
